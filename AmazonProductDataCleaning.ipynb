```python
import pandas as pd
from datetime import datetime
from datetime import date
```

```python
df = pd.read_csv('D:\data.txt',encoding='latin',sep='|')
df.head(5)
```

```python
'''Data Cleaning 
   1.remove special charaterfrom rank,price,revenu column
   2.remove space from column name
   3.weight column add 0 for null value
   4.revenue column rerplace null with 0
   5.rating couln replace null value with 0
   6.format date in YYYY-MM-DD Format
   '''
```

```python
#rename a column 
df.rename(columns={'Product Name':'ProductName','Available From':'AvailableFrom','Est. Revenue':'Revenue'},inplace=True)
```

```python
df.head(5)
```

```python
#remove special charaterfrom rank,price,revenu column
df = df.replace('\$','',regex=True)
df = df.replace('\#','',regex=True)
#df = df.astype({'Price':float,'Revenue':float})
df.head(5)
```

```python
df.isnull().sum()
```

```python
#remove unwanted row
df.dropna(subset=['Price'],inplace=True)
df.dropna(subset=['Brand'],inplace=True)
df.head(5)
```

```python
#replace null value with 0
df['Rating'].fillna(0,inplace=True)
df['Weight'].fillna(0,inplace=True)
df['Rank'].fillna(0,inplace=True)
df['Revenue'].fillna(0,inplace=True)
```
